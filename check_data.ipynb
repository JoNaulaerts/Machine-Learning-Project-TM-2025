```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load your datasets
# For UK Housing
df_housing = pd.read_csv('Dataset_1_UK_Housing/Data/price_paid_records.csv')

# For UK Electricity (you'll need to combine multiple files)
# Use glob or os.listdir to get all files
import glob
electricity_files = glob.glob('Dataset_2_UK_Historic_Electricity_Demand_Data/Data/*.csv')
df_electricity = pd.concat([pd.read_csv(f) for f in electricity_files], ignore_index=True)

# First inspection
print(df_housing.info())
print(df_housing.head())
print(df_housing.describe())
# First inspection
print(df_electricity.info())
print(df_electricity.head())
print(df_electricity.describe())
```

```python
# Remove spaces and make lowercase
df_housing.columns = df_housing.columns.str.lower().str.replace(' ', '_')
df_electricity.columns = df_electricity.columns.str.lower().str.replace(' ', '_')
```
